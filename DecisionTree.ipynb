```python
# This Model Uses a Decision Tree Classifier to predict whether or not customers will pay off their loans and whether or not to accept loan requests
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
data = pd.read_csv("data/Decision_Tree_Dataset.csv", sep=",", header=0)
data.head()
```

```python
print("Dataset length:", len(data))
print("Dataset shape:", data.shape)
```

```python
data.drop(columns=['sum'], inplace=True)
data
```

```python
data.rename(columns={'1': 'Initial Payment', '2': 'Last Payment', '3':'Credit Score', '4':'House Number', 'Unnamed: 5':'Result'}, inplace=True)
```

```python
data
```

```python
X = data.values[:, 0:4]
y= data.values[:, 4]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
model = DecisionTreeClassifier(criterion='entropy', random_state=2, max_depth=3, min_samples_leaf=5)
model.fit(X_train, y_train)
```

```python
y_prediction = model.predict(X_test)
y_prediction
```

```python
accuracy = accuracy_score(y_test, y_prediction)
print("Total Accuracy:", (accuracy*100))
```
